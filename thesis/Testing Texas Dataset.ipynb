```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error

x_train, x_test, y_train, y_test = train_test_split(df['loc.var'], df['sd.sadness'], test_size=0.2, train_size=0.8, random_state = 0)
```

```python
#oversample = SMOTE()
x_str = "loc.var"
y_str = "mean.sadness"

x = df[x_str].to_numpy().reshape(-1, 1)
y = df[y_str].to_numpy().reshape(-1, 1)
#x, y = oversample.fit_resample(x, y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, train_size=0.8, random_state = 0)


linearModel = LinearRegression()
x_train= x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
linearModel.fit(x_train, y_train)
y_pred = linearModel.predict(x_test)
print("Linear Regression model performed on " + x_str + " vs. " + y_str)
print("\nCorrelation Coefficient = " + str(linearModel.coef_))
print("R^2 score = " + str(linearModel.score(x, y)))
print("\nMean Squared Error = " + str(mean_squared_error(y_test, y_pred)))
#print(linearModel.intercept_, linearModel.coef_, linearModel.score(x, y))
```
